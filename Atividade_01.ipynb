2. Explique o conceito de conjunto de treinamento, conjunto de validação e conjunto de teste em machine learning. 

Para que um projeto de ML seja desenvolvido, um conjunto de dados orignal deve ser dividido em subconjuntos, são eles: (i) treinamento, (ii) validação e (iii) teste. Cada um deles tem uma característica e objetivo diferentes.

(i) Treinamento

É a maior fatia dos dados e tem cono objetivo ensinar o modelo.

(ii) Validação 

É uma porção menor (10-20%) separada do treinamento e usado durante o desenvolvimento do modelo para tomar decisões, mas sem que o modelo aprenda diretamente com esse conjunto de dados.

(iii) Teste

Pedaço do dataset (10-20%) que é reservado para avaliar o desempenho final do modelo. É intocado até o fim do treinamento e simula o uso do medlo em uma situação real.

Caso essa divisão não seja adotada, o modelo pode simplesmente decorar as respostas em vez de aprender, o que é chamado de overfitting (sobreajuste). Ou seja, ele vai bem nos dados de treino, entregando boas métricas de validação (acurácia, precisão e outras), mas falha em dados novos, o que o torna inútil para o mercado real.

3. Explique como você lidaria com dados ausentes em um conjunto de dados de treinamento. 

Dados ausentes são algo recorrente na rotina de trabalho de qualquer pessoa de dados, principalmente de um cientista. Pensando nisso, duas importantes considerações devem ser feitas quando se tem um conjunto de dados de treinamento em que existem valores nulos.

A primeira é observar as estatísticas iniciais dos dados, uma vez que o tipo de distribuição (normal ou assimétrica) impacta em como se lida com valores nulos.

A segunda é que, ao identificar essas estatísticas iniciais (média, mediana, desvio-padrão, variância e existência ou não de outliers), deve-se considerar o tamanho do dataset, já que se for um conjunto pequeno, retirar as linhas dos dados nulos impactaria diretamente na análise, o que difere (provavelmente) de um dataset de maior proporção.

Feita essa análise inicial, algumas são as possibilidades de substituição dos dados, considerando as medidas separatrizes e suas características. Vamos a elas.

Antes, vale um adendo em relação a média e mediana.

A média é uma medida extermamente sensível a outliers (valores discrepantes) pela sua própria fórmula (n1 + n2 + n... / n). Isso a torna um parâmetro vulnerável de ser adotado dentro de um conjunto de dados assimétrico, pois ela "puxa" para os extremos dos seus outliers e distorce a realidade dos dados. Ou seja: ela te faz acredita em algo que não é o real e te faz concluir uma análise equivocada.

Já a mediana é o contrário. Por ser uma medida cujo valor é o centro exato do conjunto dos dados, torna-se um parâmetro mais adequado e menos sensível a outliers. Assim, é mais confiável para ser uma substituição possível em datasets que possuem distribuição assimétrica.

---- DADOS SIMÉTRICOS ----

Escolher a média ou a mediana pode ser quase invisível, já que os valores são simétricos (distribuição normal) e a diferença real será quase imperceptível. Desa forma, pode-se escolher qualquer das medidas para esse caso.

---- DADOS ASSIMÉTRICOS ----

Aqui está a diferença do analista/cientista que sabe ou não aplicar os conceitos estatísticos. 

Se for aplicada a média a um conjunto assimétrico, o impacto será gigantesco. Isso porque esse valor não será o que os dados realmente refletem, mas sim exceções que trazem para si o valor dessa medida.

Nesse caso, o mais adequado é adotar a mediana, já que menos sensível a outliers e mais fidedigna à realidade dos dados.

4. O que é uma matriz de confusão e como ela é usada para avaliar o desempenho de um modelo preditivo? 

Matriz de confusão é uma tabela que organiza as previsões de um modelo de classificação comparando o que foi previsto com o que realmente aconteceu.

Trata-se de base de cálculo para métricas de avaliação dos modelos de classificação. Assim, ela é utilizada a partir de parâmetros matemáticos de acertos e erros, para que métodos de validação/avaliação dos modelos sejam desenvolvidos.

Estrutura básica (caso binário)

Tabela 2 x 2, em que os cabeçalhos das colunas são previstos positivo e negativo, respectivamente, e as linhas são real positivo e negativo, respectivamente. 
- Real positivo e previso positivo: verdadeiro positivo (VP)
- Real positivo e previsto negativo: falso negativo (FN) 
- Real negativo e previsto positivo: falso positivo (FP)
- Real negativo e previsto negativo: verdadeiro negativo (VN)

Com essa estrutura é possível encontrar métricas e avaliar o modelo a partir de cálculos matemáticos que originam métricas derivadas dessa análise, como acurácia, precisão, F1 score e recall.
