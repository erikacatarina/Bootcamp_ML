2. Escreva uma função que receba uma lista de números e retorne outra lista 
com os números primos presentes

```python
def prime_number_list(lista) -> List:
    prime_number = []
    for num in lista:
        if num > 1:
            is_prime = True
            for divisor in range(2, num):
                if num % divisor == 0:
                    is_prime = False
                    break
            if is_prime:
                prime_number.append()
    return prime_number
```

3. Escreva uma função que receba duas listas e retorne outra lista com os 
elementos que estão presentes em apenas uma das listas. 

```python
def only_in_one_list(list1, list2) -> List:
    new_list = []
    for num in list1:
        if not(num in list2):
            new_list.append(num)
    for num in list2:
        if not (num in list1):
            new_list.append(num)
    return new_list
```

4. Dada uma lista de números inteiros, escreva uma função para encontrar o 
segundo maior valor na lista. 

```python
def finding_max_value(lista) -> List:
    unique = set(lista)
    if len(unique) > 2:
        sorted_list = sorted(unique)
        return sorted_list[-2]
```

6. Como identificar e tratar outliers em uma coluna numérica usando desvio 
padrão ou quartis?

```python
import pandas as pd

df = pd.DataFrame({'valores': [10, 12, 11, 13, 12, 100, 11, 9]})

# --- Método 1: Desvio padrão ---
media = df['valores'].mean()
desvio = df['valores'].std()
limite_superior = media + 3 * desvio
limite_inferior = media - 3 * desvio

outliers_std = df[(df['valores'] > limite_superior) | (df['valores'] < limite_inferior)]

# --- Método 2: Quartis (IQR) ---
Q1 = df['valores'].quantile(0.25)
Q3 = df['valores'].quantile(0.75)
IQR = Q3 - Q1

limite_inf_iqr = Q1 - 1.5 * IQR
limite_sup_iqr = Q3 + 1.5 * IQR

outliers_iqr = df[(df['valores'] < limite_inf_iqr) | (df['valores'] > limite_sup_iqr)]

# Tratar: remover outliers
df_limpo = df[(df['valores'] >= limite_inf_iqr) & (df['valores'] <= limite_sup_iqr)]

print("Outliers (IQR):\n", outliers_iqr)
print("Dados limpos:\n", df_limpo)
```

7. Como concatenar vários DataFrames (empilhando linhas ou colunas), mesmo que tenham colunas diferentes? 

Dica: Utiliza-se pd.concat() especificando axis=0 (linhas) ou axis=1 (colunas). 
Quando há colunas diferentes, os valores ausentes são preenchidos com NaN. 

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'C': [7, 8]}) 

# Empilhar linhas (axis=0) - colunas que não existem viram NaN
resultado_linhas = pd.concat([df1, df2], axis=0)
print(resultado_linhas)

# Empilhar colunas (axis=1) - alinha pelos índices das linhas
resultado_colunas = pd.concat([df1, df2], axis=1)
print(resultado_colunas)
```

9. Utilizando pandas, como selecionar uma coluna específica e filtrar linhas

10. Utilizando pandas, como lidar com valores ausentes (NaN) em um 
DataFrame?

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'A': [1, np.nan, 3], 'B': [4, 5, np.nan]})

# 1. Identificar onde há NaN
print(df.isna())          # True/False por célula
print(df.isna().sum())    # quantidade de NaN por coluna

# 2. Remover linhas com qualquer NaN
df_sem_nan = df.dropna()

# 3. Preencher NaN com um valor (ex: média da coluna)
df_preenchido = df.fillna(df.mean(numeric_only=True))

# 4. Preencher com valor fixo
df_zerado = df.fillna(0)

print(df_sem_nan)
print(df_preenchido)
```
